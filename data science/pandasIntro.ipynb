##  Manipulating Data

```python
#Renaming Columns
df.rename(columns = {'OdName':'Country','AreaName' : 'Continent','DevName' : 'DevStatus'},inplace = True)
df.head()
#inplace = True   bring changes in df also
```

```python
df.Type.unique()
```

```python
df.drop(columns = ['Type','Coverage','AREA','DEV'], inplace = True)

#if the runs one time then the value get removed, if run again, if will give error
```

```python
#Find the number of immigrants from India and China and Afghanistan in year 1990 - 2005

df.loc[(df['Continent'] == 'Asia') & (df['DevStatus'] == 'Developing regions') ]  #for condition in multiple column
```

```python
#for matching multiple values
countries = ['India','China','Afghanistan']
df.loc[df['Country'].isin(countries)]
```

```python
#soring DataFrame
# df.sort_values(2004)           #ascending order
df.sort_values(2004 , ascending = False)           #descending order
```

```python
#find top 10 countries in 2013.
df.sort_values(2013 , ascending = False).head(10)[['Country',2013]]
```

```python
#Find top 5 countries of Asia in year 2008
conti = df.loc[(df['Continent'] == 'Asia') ]
final = conti.sort_values(2008,ascending= False).head()
final[['Continent','Country',2008]]
```
